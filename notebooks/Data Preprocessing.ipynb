```python
import gzip
import shutil
import csv
import pandas as pd
```

```python
# Converting this gz file into txt then csv and then adding column heading as per data/README.txt file 
# for dev.txt

# Input and output file paths
input_gz_file = '../data/hindi-visual-genome-dev.txt.gz'
output_txt_file = '../data/dev.txt'

# Convert .gz to .txt
with gzip.open(input_gz_file, 'rt', encoding='utf-8') as f_in:
    with open(output_txt_file, 'w', encoding='utf-8') as f_out:
        shutil.copyfileobj(f_in, f_out)

print("Conversion complete.")


# Input and output file paths
input_txt_file = '../data/dev.txt'
output_csv_file = '../data/output_dev.csv'

# Read tab-separated lines and write as CSV
with open(input_txt_file, 'r', encoding='utf-8') as txt_file:
    with open(output_csv_file, 'w', encoding='utf-8', newline='') as csv_file:
        reader = csv.reader(txt_file, delimiter='\t')
        writer = csv.writer(csv_file)
        for row in reader:
            writer.writerow(row)

print("Conversion to CSV complete.")


df = pd.read_csv("../data/output_dev.csv" , names=['image_id' , 'x' , 'y' , 'width' , 'height' , 'english_text' , 'hindi_text'] , header=None)
df.to_csv(output_csv_file, index=False)
print(df.shape)
df.head()
```

# Merging all csv files

```python
df1 = pd.read_csv("../data/output_dev.csv")
df2 = pd.read_csv("../data/output_train.csv")
df3 = pd.read_csv("../data/output_test.csv")

print("_________shapes________")
print(f"for dev -----> {df1.shape}")
print(f"for train -> {df2.shape}")
print(f"for test ---> {df3.shape}")
```

```python
# Vertically concatenate the DataFrames
merged_df = pd.concat([df1, df2, df3], ignore_index=True)

# Optional: View result
print(merged_df.shape)
merged_df.head()
```
